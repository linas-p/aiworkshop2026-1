```python
import pandas as pd
```

```python
df_ = pd.read_csv("ldia.csv", sep=";")
df_.head()
```

```python
print(df_['Answer'][0])
```

```python
prompt_template = """<|start|>system<|message|>You are ChatGPT, a large language model trained by OpenAI.\nKnowledge cutoff: 2024-06\nCurrent date: 2025-08-05\n\nReasoning: medium\n\n# Valid channels: analysis, commentary, final. Channel must be included for every message.<|end|>
<|start|>user<|message|>
{}
<|end|>
<|start|>assistant<|channel|>final<|message|>
{}
<|end|><|return|>
"""
```

```python
# create text column for training
df_['text'] = df_.apply(
    lambda row: prompt_template.format(row['Question'], row['Answer']),
    axis=1
)
```
