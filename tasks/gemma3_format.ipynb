```python
import pandas as pd
```

```python
df_ = pd.read_csv("ldia.csv", sep=";")
df_.head()
```

```python
print(df_['Answer'][0])
```

```python
prompt_template = """<start_of_turn>user
{}
<end_of_turn>
<start_of_turn>model
{}
<end_of_turn>
"""
```

```python
# create text column for training
df_['text'] = df_.apply(
    lambda row: prompt_template.format(row['Question'], row['Answer']),
    axis=1
)
```
