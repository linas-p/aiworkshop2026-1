```python
import pandas as pd
```

```python
df_ = pd.read_csv("ldia.csv", sep=";")
df_.head()
```

```python
print(df_['Question'][0])
```

```python
prompt_template = """
<|start_of_role|>user<|end_of_role|>User: 
{}
<|end_of_text|>
<|start_of_role|>assistant<|end_of_role|>#### Analysis
{}
<|end_of_text|>
"""
```

```python
# create text column for training
df_['text'] = df_.apply(
    lambda row: prompt_template.format(row['Question'], row['Answer']),
    axis=1
)
```
